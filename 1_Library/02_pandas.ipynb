```python
# read from csv file
stu_info = pd.read_csv("dataset/stu_info_2024.csv")
# print(stu_info)
df_stu_info = pd.DataFrame(stu_info)
print(df_stu_info)
```

```python
print(df_stu_info.info())
```

```python
# basic statistics
print(df_stu_info.describe())
```

```python
print(df_stu_info.describe(include="all"))
```

```python
# Fech one of column in dataframe
print(df_stu_info["Name"])
```

```python
print(df_stu_info[["Name", "Score"]])
```

```python
# filter or search into you'r data
print(df_stu_info[df_stu_info["Score"] > 80])
```

```python
# add a new column
df_stu_info["Passed"] = df_stu_info["Score"] > 80
print(df_stu_info)
```

```python
df_stu_info["description"] = "-"
print(df_stu_info)
```

```python
# drop a column
"""
axis = 1 => Column
axis = 0 => Row
"""
df_stu_info = df_stu_info.drop("description", axis=1)
print(df_stu_info)
```

```python
# sort
print(df_stu_info.sort_values(by="Score", ascending=False))
```

```python
# Average score by age
print(df_stu_info.groupby("age")["Score"].mean())
```

```python
df2 = df_stu_info.copy()
# print(df2)
# df2.loc[1, "Score"] = np.nan # loc => location
df2.loc[1, "Score"] = 96 # loc => location
print(df2)
```

```python
df2.fillna(df2["Score"].mean(), inplace=True) # fill NaN with mean
df2
```

```python
df_clean = df2.dropna()
print("\n DataFrame after dropping rows with NaN:")
df_clean
```

```python
data_01 = pd.DataFrame({
    "Name": ["Alice", "Jack", "rock"],
    "Age": [24,25,25],
    "Score": [88,92,46]
    })

data_extra = pd.DataFrame({
    "Name": ["Alice", "Jack", "rock"],
    "City": ["tbz", "urm", "shir"],
})

merged = pd.merge(data_01, data_extra)
print("\n----------- data 01 -----------")
print(data_01)
print("\n----------- data extra --------")
print(data_extra)
print("\n----------- merged ------------")
print(merged)
```

```python
data_02 = pd.DataFrame({
    "Name": ["Alice", "Jack", "bob", "ros"],
    "Age": [24,25,25,27],
    "Score": [88,92,46,79]
    })

data_extra_02 = pd.DataFrame({
    "Name": ["Alice", "Jack", "bob"],
    "City": ["tbz", "urm", "shir"],
})

merged_02 = pd.merge(data_02, data_extra_02)
print("\n----------- data 01 -----------")
print(data_02)
print("\n----------- data extra --------")
print(data_extra_02)
print("\n----------- merged ------------")
print(merged_02)
```
