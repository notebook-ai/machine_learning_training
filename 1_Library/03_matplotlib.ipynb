```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sample one

x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.title("Simple line plot")
plt.xlabel("X")
plt.ylabel("sin(x)")
plt.show()
```

```python
plt.plot(x, np.sin(x), label = "sin(x)")
plt.plot(x, np.cos(x), label = "cos(x)")
plt.legend()
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [10,15,7,12]
plt.bar(categories, values)
plt.title("Bar plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
# print(x)
# print(y)
plt.scatter(x,y)
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='darkblue')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,4))
x = np.random.rand(50)
y = np.random.rand(50)
axs[0].plot(x, np.sin(x)); axs[0].set_title("sinos")
axs[1].plot(x, np.cos(x)); axs[1].set_title("cosinos")
plt.show()
```

```python
plt.barh(categories, values)
plt.title("Bar plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
colors = np.random.rand(50)
sizes = np.random.rand(50)  * 300

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.5, cmap="viridis")
plt.colorbar()
plt.show()
```

```python
epochs = np.arange(1,11)
train_loss = np.random.rand(10)
val_loss = np.random.rand(10)

plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, val_loss, label = "validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
