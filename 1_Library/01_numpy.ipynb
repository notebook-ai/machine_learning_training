```python
# 1 dimension
arrary_1 = np.array([1,2,3,4,5,6]) 
print(arrary_1)
print("Dimensions in arrary_1 are: ", arrary_1.ndim)
print("shape arrary_1:", arrary_1.shape)
```

```python
# 2 dimensions
array_2 = np.array(
    [
        [1,2,3],
        [4,5,6]
    ]
) 
print(array_2)
print("Dimensions in array_2 are: ", array_2.ndim)
print("shape array_2:", array_2.shape)
```

```python
# 3 dimensions
array_3 = np.array(
    [
        [
            [1, 2, 3], 
            [4, 5, 6]
        ], 
        [
            [1, 2, 3], 
            [4, 5, 6]
        ]
    ])
print(array_3)
print("Dimensions in array_3 are: ", array_3.ndim)
print("shape array_3:", array_3.shape)
```

```python
# 5 dimension
array_5 = np.array([1,2,3,4], ndmin=5)
print(array_5)
print("Dimensions in array_5 are: ", array_5.ndim)
print("shape array_5:", array_5.shape)
```

```python
# Feature matriz: 5 sample , each with [height, weight]
data = np.array(
    [
        [170, 65],
        [180, 80],
        [160, 55],
        [175, 70],
        [165, 60],
    ]
)
print("Features:\n", data)
print("\nDimensions in data are: ", data.ndim)
print("shape:", data.shape)
```

```python
# Mean height and mean weight
# Mean => sum all numbers, then divid by count of numbers
print("Mean Height:", np.mean(data[:,0])) # coulmn 0
print("Mean weight:", np.mean(data[:,1])) # coulmn 1

print("Mean Height(custome):", np.mean(data[0:2,0])) # row effected has: 0,1 
```

```python
# Generate random dataset
# 100 sample, 3 feature (values between 0 and  1)
random_dataset = np.random.rand(100, 3)
print("Dimensions in random_dataset are: ", random_dataset.ndim)
print("shape random_dataset:", random_dataset.shape)

print(random_dataset[:5])
```

```python
# Multiplying two matrices
a = np.array(
    [
        [1,2],
        [3,4]
    ]
)
b = np.array(
    [
        [5,6],
        [7,8]
    ]
)
result = a*b
print(result)
```

```python
# Tranpose  of matrix
# swap rows and columns
array = np.array(
    [
        [1,2,3],
        [4,5,6]
    ]
) 
print("Original shape of array:", array.shape)
print("Tranpose shape of array:", array.T.shape)
```

```python
# softmax function (classification)
# activation function in nerual network

z = np.array([2.0,1.0,0.1])
softmax = np.exp(z) / np.sum(np.exp(z))
print("Softmax result: ", softmax)
```

```python
# Mean Squared Error (MSE)
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])
mse = np.mean((y_true - y_pred) ** 2)
print("MSE:", mse)
```
