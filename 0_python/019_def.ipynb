```python
# recursive function
number = int(input("Enter a number: "))
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
    
result = factorial(number)
print(f"{number} factorial is: {result}")
```

```python
def square_calc(lst_nums):
    lst_new_nums = []
    for n in lst_nums:
        lst_new_nums.append(n*n)
    return lst_new_nums

numbers = [1,4,7]
squared_numbers_01 = square_calc(numbers)
print(squared_numbers_01)
```

```python
def apply_function(lst, func):
    return [func(x) for x in lst]

def square(x):
    return x * x

numbers = [1,2,3,4,5]
squared_numbers = apply_function(numbers, square)
print(squared_numbers)
```

```python
def calculator(operation, x, y):
    if operation =="+":
        return x + y
    elif operation =="-":
        return x - y
    elif operation == "*":
        return x * y
    elif operation == "/":
        if y!=0:
            return x / y
        else:
            return "Cannot divided by zero"
    elif operation == "%":
        return x % y
    else:
        return "Invalid operation"
    
print(calculator("+", 5,9))
print(calculator("+", 6,2))
print(calculator("*", 7,3))
print(calculator("/", 79, 9))
print(calculator("/", 88, 0))
print(calculator("%", 359, 73))

```

```python
def avg_numbers(nn):
    total = sum(nn)
    count = len(nn)
    return total / count

nn = [75, 983, 101, 1015]
print(avg_numbers(nn))
```

```python
def find_common_elements(lst1, lst2):
    lst_common = [item for item in lst1 if item in lst2]
    return lst_common

list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]
print(find_common_elements(list1, list2))
```

```python
def find_common_elements_02(lst1, lst2):
    lst_common_02 = []
    for item in lst1:
        if item in lst2: 
            lst_common_02.append(item)
    return lst_common_02

list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]
print(find_common_elements_02(list1, list2))
```

```python
def revers_string(text):
    return text[::-1] # start = first, step = last, step = -1
text = "laptop"
print(revers_string(text))   
```

```python
# way two
def revers_string_02(text):
    new_text = ""
    i = 1
    while True:
        try:
            new_text += text[-i]
            i += 1
        except IndexError:
            break
    return new_text

text = "laptop"
print(revers_string_02(text))
```

```python
def password_manager():
    stored_password = "secret"
    # with open("notes/pass.txt", 'r') as file:
    #     stored_password = file.read()

    # file = open("notes/pass.txt", "r")
    # stored_password = file.read()

    def check_pass(input_password):
        if input_password == stored_password:
            return True
        else:
            return False
        
    
    def update_pass(old_password, new_password):
        nonlocal stored_password
        if check_pass(old_password):
            stored_password = new_password
            return "password updated successfuly"
        else:
            return "You'r old password is incorrect password"
    
    return check_pass, update_pass
```

```python
password_manager()
```

```python
check_password, update_password = password_manager() # check for the main function, update for the second
print(check_password("test")) # output: True
print(check_password("secret.")) # output: False
```
