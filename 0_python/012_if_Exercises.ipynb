```python
# The Number is positive, negetive or zero
a = int(input("you'r number: "))
if a > 0:
    print("you'r number is positive")
elif a < 0: 
    print("you'r number is Negetive")
else:
    print("you'r number is Zero")
```

```python
# temperature in degrees Celsius
tem = int(input("Enter temperature:"))
if tem < 0:
    print("Freezing")
elif tem >= 0 and tem <=10:
    print("Cold")
elif tem >=11 and tem <=20:
    print("Cool")
elif tem >=21 and tem <=30:
    print("Warm")
else:
    print("Hot")    
```

```python
# BMI => body mass index
w = int(input("you'r weight:")) # kg
h = float(input("you'r height:")) # meter

BMI = (w / (h ** 2))
#print(BMI)

if BMI < 18.5:
    print(f"you'r BMI is {BMI:.2f}, result: Underweight")
elif BMI >=18.5 and BMI <= 24.9:
    print(f"you'r BMI is {BMI:.2f}, result: Normal weight")
elif BMI >= 25 and BMI <= 29.9:
    print(f"you'r BMI is {BMI:.2f}, result: Over weight")
else:
    print(f"you'r BMI is {BMI:.2f}, result: Obese")
```
