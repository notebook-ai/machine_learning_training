```python
# convert a list of string to integer
def convert(lst_str_nums):
    new_nums = []
    for item in lst_str_nums:
        new_nums.append(int(item))
    return new_nums

lst_str_nums = ["1", "2", "3", '4']
print(lst_str_nums)
print(convert(lst_str_nums))
```

```python
def convert_02(lst_nums_02):
    valid_items = []
    invalid_items = []
    for item in lst_nums_02:
        try:
            valid_items.append(int(item))
        except ValueError:
            invalid_items.append(item)
    data = {
        "Numbers": valid_items,
        "Others": invalid_items
    }
    return data
lst_nums_02 = ["503", "2074", "42", "four", "+", "89"]
print(lst_nums_02)
result = convert_02(lst_nums_02)

print(result)
```

```python
vowels = ["a", "e", "i", "o", "u"]
def remove_vowels(text):
    new_text = ""
    for letter in text:
        if letter not in vowels:
            new_text += letter
    return new_text

text = input("Enter a word: ")
print(text)
print(remove_vowels(text))
```

```python
def count_occurrences(nums):
    data = {}
    cnt_item = 0
    lenght = len(nums)
    i = 0
    for item in nums:
        while i < lenght:
            if item == nums[i]:
                cnt_item += 1
            i += 1
        data[str(item)] = cnt_item
        cnt_item = 0
        i = 0
    return data

all_nums = [1,2,2,9,7,2,7,6,1,7,5,1]
print(all_nums)
print(count_occurrences(all_nums))
```
