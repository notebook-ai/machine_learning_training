```python
"""
Sets are used to store multiple items in a single variable.
Set is one of 4 built-in data types in Python used to store collections of data,
the other 3 are List, Tuple, and Dictionary, all with different qualities and usage.
"""
```

```python
nums_01 = {1,2,3,4,5}
nums_02 = set([1,2,3,4,5])
print(type(nums_01))
print(type(nums_02))

print(nums_01)
print(nums_02)
```

```python
colors = {"Red", "Green", "Blue"}
print(colors)
```

```python
# Adding element
colors.add("purple")
print(colors)
```

```python
# Remove a element
colors.remove("Green")
print(colors)
```

```python
# discarding an element
colors.discard("Green")
print(colors)
```

```python
# Popping an elemet
popped_element = colors.pop() # get first element then remove that
print(popped_element)
print(colors)
```

```python
# Clear the set
colors.clear()
print(colors)
```

```python
# Union: Returns a new set with elemnts from both sets
"""
If they have a common value, only show one instance in the output.
"""
odd_nums = {1,3,5,7,9,11}
even_nums = {0,2,4,6,8,10}
result_01 = odd_nums.union(even_nums)
result_02 = even_nums.union(odd_nums)
print(result_01)
print(result_02)
```

```python
set1 = {1,2,3}
set2 = {3,4,5}
```

```python
res = set1.union(set2)
print(f"set_one: {set1}")
print(f"set_two: {set2}")
print(f"resutl is: {res}")
```

```python
# Intersection: return a new set with elements common values to both sets
intersection_set = set1.intersection(set2)
print(f"set_one: {set1}")
print(f"set_two: {set2}")
print(f"intersection_set: {intersection_set}")
```

```python
diff_set_01 = set1.difference(set2)
diff_set_02 = set2.difference(set1)
print(f"set_one: {set1}")
print(f"set_two: {set2}")
print(f"difference_one_to_two: {diff_set_01}")
print(f"difference_two_to_one: {diff_set_02}")
```

```python
# symmetric difference
sym_diff_set = set1.symmetric_difference(set2)
print(f"set_one: {set1}")
print(f"set_two: {set2}")
print(f"symmetric difference: {sym_diff_set}")
```

```python
# update: update the set with elements from the other set
set1.update(set2)
print(f"set_one: {set1}")
```

```python
set_03 = {1,2}
set_04 = {1,2,3,4}

print(f"is_subset: {set_03.issubset(set_04)}")
print(f"is_superset: {set_04.issuperset(set_03)}")
print(f"is_disjoint: {set_03.isdisjoint({3,4,5})}")
```
