```python
# check odd or even number for numbers greater than 9

num = int(input("please eneter you'r number:"))
if num > 9 :
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
else:
    print("you'r number is not greater than 9")
```
