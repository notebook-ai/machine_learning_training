```python
"""
3. Count vowels in a string 
etc:  a, e, i, o ,u
"""
vowels = ('a', 'e', 'i', 'o' ,'u')
cnt_vl = 0 #count_of_vowels_letter
text = "Programing in python is fun and educational."
#text = "python"
for letter in text:
    if letter in vowels:
        cnt_vl +=1

print(f"The text are: {text}")
print(f"Count of vowels letter: {cnt_vl}")
```

```python
"""
2. generate a multiplication table for numbers from 1 to 10
"""
for i in range(1,11):
    print('')
    for j in range(1,11):
        print(i * j, end=' ')
    print()
```

```python
"""
2. Other way to generate a multiplication table for numbers from 1 to 10
"""
tbl_mul = {}
new_line = []
i = 0
j = 0
for i in range(1,11):
    for j in range(1,11):
        new_line.append(i * j)
    # print(new_line)
    tbl_mul[i] = list(new_line)
    new_line.clear()

#print(tbl_mul) 
for item in range(1,11):
    for row in tbl_mul[item]:
        print(row, end=" ")
    print()
```
