```python
# _____operatrors_____
"""
+ (Addition): x + y
- (subtraction): x - y
* (Multiplication): x * y
/ (division): x / y
% (Modulus): x % y
** (Exponentiation): x ** y
// (floor division): x // y
"""
x = 6
y =  2
addition = x + y
subtraction = x - y
multiplication = x * y
division = x / y
modulus = x % y
exponentiation = x ** y
floor_division = x // y

print("result addition: " , addition)
print("result subtraction: " , subtraction)
print("result multiplication: " , multiplication)
print("result division: " , division)
print("result modulus: " , modulus)
print("result exponentiation: " , exponentiation)
print("result floor_division: " , floor_division)
```
