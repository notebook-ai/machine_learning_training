```python
"""
List:
*. Ordered, mutable collections of itmes
1. append(x)
2. extend(iterable)
3. insert(i, x)
4. remove(x)
5. pop([i])
6. clear()
7. index()
8. count(x)
9. sort
10. reverse
11. copy()
"""
```

```python
# creating a list
lst_nums = [1,2,3,4,5]

#Accessing elemts
print(lst_nums)
print(lst_nums[0])
print(lst_nums[2])
print(lst_nums[4])
print(lst_nums[-1])
```

```python
# Slicing lists
print(lst_nums)
print(lst_nums[1:3]) # start from index 1 To index 3, except show index 3
print(lst_nums[:2]) # start from index 0 To index 2
print(lst_nums[2:]) # start from index 2 To End_of_index
```

```python
lst_other = [1,2,3, 'a', 'b', 'c']
print(lst_other)
```

```python
# changing elements
lst_other[1] = 'ch_new_value'
print(lst_other)
```

```python
# append a single element
print(lst_other)
lst_other.append('d')
print(lst_other)
```

```python
# extend th list with multiple elements
lst_other.extend(['e', 'f'])
print(lst_other)
lst_alphabet = ['g', 'h']
lst_other.extend(lst_alphabet)
print(lst_other)
```

```python
# insert an element at a specific position
lst_other.insert(3, "inserted_value")
lst_other
```

```python
# remove a specific element by value
lst_other.remove('a')
lst_other
```

```python
# pop an element by index
# it's look like remove 
print(lst_other)
poped_element = lst_other.pop(3)
print(poped_element)
print(lst_other)
```

```python
# clear
lst_test = [1,2,3]
print(lst_test)
lst_test.clear()
print(lst_test)
lst_values = [6,5,4,3,2,1,0]
lst_test.extend(lst_values)
print(lst_test)
```

```python
# sort
lst_alphabets = ['e', 'r','a', 'w']
print(lst_alphabets)
lst_alphabets.sort()
print(lst_alphabets)
```

```python
# show index
print(lst_alphabets.index('r')) # get value then show index
```

```python
# reverse
print(lst_nums)
lst_nums.reverse()
print(lst_nums)
```

```python
# copy
lst_nums = lst_nums.copy()
print(lst_nums)
```

```python
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
print(matrix)
print(matrix[1][2]) # row 1 , column 2
print(matrix[1]) # row 1 , column 2
```

```python
# another way to reverse
lst_even_nums = [0,2,4,6,8,10]
print(lst_even_nums)
print(lst_even_nums[::-1])
```

```python
# another way to concatebating
lst_one = [1,2,3]
lst_two = [4,5,6]
print(lst_one + lst_two)
```

```python
# another way to copy
lst_three = lst_one[:]
print(lst_three)
```

```python
# finding the sum, min, max
lst_numbers = [10,20,30,40]
total = sum(lst_numbers)
minimum = min(lst_numbers)
maximum = max(lst_numbers)

print(total)
print(minimum)
print(maximum)

# check for membership
is_member = 98 in lst_numbers
not_member = 3 not in lst_numbers
print(is_member)
print(not_member)
```

```python
# repeating list
repeating_lst = lst_numbers * 2
print(repeating_lst)
```

```python
# lambda
lst_original = [1,2,3]
lst_new = list(map(lambda x:x*2, lst_original))
print(lst_original)
print(lst_new)
```
