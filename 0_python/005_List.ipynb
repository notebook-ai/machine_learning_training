```python
# creating a list
lst_nums = [1,2,3,4,5]

#Accessing elemts
print(lst_nums)
print(lst_nums[0])
print(lst_nums[2])
print(lst_nums[4])
print(lst_nums[-1])
```

```python
# Slicing lists
print(lst_nums)
print(lst_nums[1:3]) # start from index 1 To index 3, except show index 3
print(lst_nums[:2]) # start from index 0 To index 2
print(lst_nums[2:]) # start from index 2 To End_of_index
```

```python
lst_other = [1,2,3, 'a', 'b', 'c']
print(lst_other)
```

```python
# changing elements
lst_other[1] = 'ch_new_value'
print(lst_other)
```

```python
# append a single element
print(lst_other)
lst_other.append('d')
print(lst_other)
```
