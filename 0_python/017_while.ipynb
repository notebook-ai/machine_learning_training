```python
# Iterating numbers from 1 to 10
i = 1
while (i <= 4):
    print(i)
    i+=1
```

```python
# sum even nums from 1 to 10 
sum = 0
i = 0
while (i <= 9):
    if i % 2 == 0:
        sum += i
    i+=1
print(f"sum even nums from 1 to 10: {sum}")
```

```python
# mul all nums from 1 to 10
mul = 1
i = 1
while (i <= 9):
    mul *= i
    i+=1
print(f"mul even nums from 1 to 10: {mul}")
```

```python
counter_starts = 1
i = 1
j = 1
text = ""
while (i <= 10):
    while ( j <= counter_starts):
        text += "* "
        j += 1
    print(text)
    i += 1
    counter_starts = i
    j = 1
    text = ""

```
