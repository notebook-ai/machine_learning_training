```python
# tuple in python are immutable sequences
tpl_nums = (1,2,3,4,5) 
tpl_nums
```

```python
# Accessing elements
print(tpl_nums[0])
print(tpl_nums[-1])
```

```python
#slicing tuples
print(tpl_nums[1:3])
print(tpl_nums[:2])
print(tpl_nums[2:])
```

```python
# count of values in tuple variable
print(tpl_nums.count(3))
```

```python
# fetch of index form get value
print(tpl_nums.index(4))
```
